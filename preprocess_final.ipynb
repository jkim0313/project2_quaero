```python
# 행 개수가 3개 이상인 기업 리스트 및 수
over_three = []
for i in company_list:
    if len(df[df['회사명']==i]) >= 3:
        over_three.append(i)
print(len(over_three))
print(over_three)
```

```python
# 행 개수가 2개 이상인 기업 리스트 및 수
over_two = []
for i in company_list:
    if len(df[df['회사명']==i]) >= 2:
        over_two.append(i)
print(len(over_two))
print(over_two)
```

```python
# 행 개수가 3개 이상인 기업 인덱스
over_three_index = []
for i in over_three:
    over_three_index.append(df[df['회사명']==i].index)
over_three_index
```

```python
# 행 개수가 2개 이상인 기업 인덱스
over_two_index = []
for i in over_two:
    over_two_index.append(df[df['회사명']==i].index)
over_two_index
```

#### 부실기업 정의 2개를 하나의 포문안에 조건문으로 한번에 하면,
#### 첫번째 조건을 무시한채 두번째 조건으로 덮어씌워지기 때문에
#### 하나씩 나눠서 지정해준다.

```python
# 3년 연속 이자보상배율이 1미만인 기업을 부실기업으로 정의
icr_count = 0
df['부실기업1'] = ''

for i in over_three_index:
    for j in i:
        if df.loc[j, '이자보상배율'] < 1:
            icr_count += 1
        else:
            icr_count = 0
                    
        if icr_count >= 3:
            df.loc[j, '부실기업1'] = 1
        else:
            df.loc[j, '부실기업1'] = 0
```

```python
for i in df.index:
    if df.loc[:, '부실기업1'] == '':
        df.loc[i, '부실기업1'] = 0
```

```python
#2년 연속 매출액이 20% 이상 감소한 기업을 부실기업으로 정의
sales_count = 0
df['부실기업2'] = ''

for i in over_two_index:
    for j in i:
        if df.loc[j, '매출액증가율'] < -20:
            sales_count += 1
        else:
            sales_count = 0
                    
        if sales_count >= 2:
            df.loc[j, '부실기업2'] = 1
        else:
            df.loc[j, '부실기업2'] = 0
```

```python
df['부실기업'] = ''
for i in df.index:
    if df.loc[i, '부실기업1'] == 1 or df.loc[i, '부실기업2'] == 1:
        df.loc[i, '부실기업'] = 1
    else:
        df.loc[i, '부실기업'] = 0
        
# 부실기업 두가지 정의 모두에 해당하는 기업의 수
df['회사명'][df['부실기업']==1].nunique()
```

```python
# 행 개수가 3개 이상인 기업 리스트 및 수
part5_over_three = []
for i in company_list:
    if len(df_part5[df_part5['회사명']==i]) >= 3:
        part5_over_three.append(i)
print(len(part5_over_three))
print(part5_over_three)
```

```python
for i in df_part5.index:
    if df_part5.loc[i, '부실기업1'] == '':
        df_part5.loc[i, '부실기업1'] = 0
```

```python
# 행 개수가 2개 이상인 기업 리스트 및 수
part5_over_two = []
for i in company_list:
    if len(df_part5[df_part5['회사명']==i]) >= 2:
        part5_over_two.append(i)
print(len(part5_over_two))
print(part5_over_two)
```

```python
# 행 개수가 3개 이상인 기업 인덱스
part5_over_three_index = []
for i in part5_over_three:
    part5_over_three_index.append(df_part5[df_part5['회사명']==i].index)
part5_over_three_index
```

```python
# 행 개수가 2개 이상인 기업 인덱스
part5_over_two_index = []
for i in part5_over_two:
    part5_over_two_index.append(df_part5[df_part5['회사명']==i].index)
part5_over_two_index
```

#### 부실기업 정의 2개를 하나의 포문안에 조건문으로 한번에 하면,
#### 첫번째 조건을 무시한채 두번째 조건으로 덮어씌워지기 때문에
#### 하나씩 나눠서 지정해준다.

```python
# 3년 연속 이자보상배율이 1미만인 기업을 발생 전년도부터 부실기업으로 정의
icr_count = 0
df_part5['부실기업1'] = ''

for i in part5_over_three_index:
    for j in i:
        if df_part5.loc[j, '이자보상배율'] < 1:
            icr_count += 1
        else:
            icr_count = 0
            
        if icr_count >= 3:
            df_part5.loc[j-1, '부실기업1'] = 1
            df_part5.loc[j, '부실기업1'] = 1
        else:
            df_part5.loc[j, '부실기업1'] = 0
```

```python
for i in df_part5.index:
    if df_part5.loc[i, '부실기업1'] == '':
        df_part5.loc[i, '부실기업1'] = 0
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE

# smote = SMOTE(sampling_strategy='minority', random_state=0)
```

```python
X = df.iloc[:, 9:].values
y = df.iloc[:, 7].values
```

```python
# 언더샘플링
undersampling = RandomUnderSampler(sampling_strategy='majority')    # "majority" => 다수 집단 클래스를 소수 집단 클래스 데이터수와 동일하게 맞춘다.
X_under, y_under = undersampling.fit_resample(X, y)
```

```python
# 오버샘플링
oversampling = RandomOverSampler(sampling_strategy='minority')    # "minority" => 소수 집단 클래스를 다수 집단 클래스 데이터수와 동일하게 맞춘다.
X_over, y_over = undersampling.fit_resample(X, y)
```

```python
# SMOTE
smote_sample = SMOTE(sampling_strategy='minority')
X_sm , y_sm = smote_sample.fit_resample(X, y)
```
